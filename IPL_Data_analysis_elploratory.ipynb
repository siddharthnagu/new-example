When a team wins after batting first 
# *italicised text*

> Indented block



````python
batting_first = data[data['win_by_runs']!=0]
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()), labels= list(batting_first['winner'].value_counts().keys()),autopct = '%0.1f%%')
plt.show()
````

When a team wins after batting second



````python
batting_second = data[data['win_by_wickets']!=0]
plt.figure(figsize=(10,10))
plt.pie(list(batting_second['winner'].value_counts()), labels= list(batting_second['winner'].value_counts().keys()),autopct = '%0.1f%%')
plt.show()
````

Winning matches after fielding first across various venues

````python
new_matches = data[data['result']== 'normal']
new_matches['win_batting_first'] = np.where((new_matches.win_by_runs > 0), 'Yes', 'No')
new_matches.groupby('venue')['win_batting_first'].value_counts().unstack().plot(kind = 'barh', stacked = True, figsize=(15,15))
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel("Frequency",fontsize = 15)
plt.ylabel("Venue", fontsize = 25)
````

man of the match most of the times


````python
plt.figure(figsize=(10,5))
sns.barplot(x = data.player_of_match.value_counts()[:10].index , y=data.player_of_match.value_counts()[:10])
plt.title("Man of the match")
plt.xlabel("player")
plt.ylabel('Match')
plt.xticks(rotation = 90)
````

Maximum number of toss won by teams

````python
plt.figure(figsize=(20,10))
plt.bar(list(data['toss_winner'].value_counts().keys()),list(data['toss_winner'].value_counts().keys()))
plt.xticks(rotation = 45)
plt.show()
````

comparing the toss decsion

````python
plt.figure(figsize=(20,10))
sns.countplot(data['season'].sort_values(), hue= data['toss_decision'], order = data['season'].value_counts().index)
plt.title("Decision to bat or field over the years")
plt.xlabel("Year")
plt.ylabel("count")
plt.show()
````
